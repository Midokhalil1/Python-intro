```python
import tensorflow_hub as hub
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
def load_image(img_path):
    img = tf.io.read_file(img_path)
    # Detects image format and converts it into Tensor
    img = tf.image.decode_image(img, channels=3)
    # convert to desired data type
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img
```

```python
content_image = load_image('content.jpeg')
style_image = load_image('style.jpeg')
```

```python
plt.imshow(np.squeeze(content_image))
plt.show()
```

```python
plt.imshow(np.squeeze(style_image))
plt.show()
```

```python
styled_image = model(tf.constant(content_image), tf.constant(style_image))
```

```python
plt.imshow(np.squeeze(styled_image))
plt.show()
```
